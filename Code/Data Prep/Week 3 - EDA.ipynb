# FIFA 19 Exploratory Data Analysis

```python
# import necessary libraries
%matplotlib inline

import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
# load the data
df = pd.read_csv('cleaned data/fifa19data_clean_no_outliers.csv')
df
```

```python
# correlation between overall rating and value
plt.plot(df['Overall'],df['Value'],'o')
```

```python
# sorted list of players by rating
sorted_overall_rating=df.sort_values('Overall',ascending=False).reset_index(drop=True)
sorted_overall_rating
```

```python
# sorted list of players by potential rating who are 21 or younger
younger_than_21 = df['Age']<21
potential_younger_than_21=df[younger_than_21]
sorted_potential_younger_than_21=potential_younger_than_21.sort_values('Potential',ascending=False).reset_index(drop=True)
sorted_potential_younger_than_21
```

```python
# club who spends the most on wages
club_with_most_wages = df.groupby(by='Club')['Wage'].sum()
club_with_most_wages.sort_values(ascending=False).head(1)
```

```python
# club with the highest rated players (average rating of top 10 players)
df.groupby(by='Club')['Overall'].apply(lambda grp:grp.nlargest(10).mean()).sort_values(ascending=False).head(1)
```

```python
# club with the highest potential (average rating of players' with the top 10 potential at each club)
df.groupby(by='Club')['Potential'].apply(lambda grp:grp.nlargest(10).mean()).sort_values(ascending=False).head(1)
```

```python
# club with the highest potential young players (players under 21)
sorted_potential_younger_than_21.groupby(by='Club')['Potential'].apply(lambda grp:grp.nlargest(10).mean()).sort_values(ascending=False).head(1)
```

```python
# each country's average rating of top 10 outfield players and the top rated GK
no_Goalie = df['Position']!='GK'
df_No_Goalie=df[no_Goalie]
df_No_Goalie.groupby(by='Nationality')['Overall'].apply(lambda grp:grp.nlargest(10).mean())
```

```python
only_Goalie = df['Position']=='GK'
df_Only_Goalie = df[only_Goalie]
df_Only_Goalie.groupby(by='Nationality')['Overall'].max()
```

```python
# average rating of top 5 players in each position
df.groupby(by='Position')['Overall'].apply(lambda grp:grp.nlargest(5).mean()).sort_values(ascending=False)
```

```python
# potential vs age 
plt.xlabel('Age')
plt.ylabel('Average Potential')
plt.plot(df.groupby('Age')['Potential'].mean().round())
```

```python
# overall rating vs age
plt.xlabel('Age')
plt.ylabel('Average Overall Rating')
plt.plot(df.groupby('Age')['Overall'].mean().round())
```

```python
# Is overall rating a good indicator of value?
plt.plot(df['Overall'], df['Value'],'o')
```

```python
# Which player is potentially undervalued? Relationship between potential rating and current value...
# (-ve correlation maybe?)
plt.plot(df['Potential'], df['Value'],'o')
```

```python
# As seen from the above graph, there are a lot more people with high potential but with very low salary. 
```

```python
# Identify undervalued players...so those players with the largest overall rating per million euro 
# (overall rating / value)
df.sort_values(by=['Value','Overall'],ascending=[True,False])
```

```python
# Identify players who have the potential to be of good value...
df.sort_values(by=['Value','Potential'],ascending=[True,False])
```

```python
# young players with the largest potential rating per million euro
potential_younger_than_21.sort_values(by=['Value','Potential','Age'],ascending=[True,False,True])
```
